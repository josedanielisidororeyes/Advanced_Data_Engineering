```python
# Carga de librerías

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import nltk
from nltk.corpus import stopwords
import pandas  as pd
```

# Ejercicio 1

## Creación de CountVectirizer y TdifVectorizer

```python
corpus_es = [
    "Me encantó la película, los actores fueron increíbles",
    "No me gustó la película, el guion fue muy malo",
    "La actuación fue excelente, pero la historia era predecible",
    "Película aburrida, me dormí a la mitad",
    "¡Maravillosa! Recomiendo esta película a todos mis amigos",
    "El guion estaba mal escrito, pero la actuación salvó la película",
    "Demasiado larga y lenta, no la volvería a ver",
    "Me gustó mucho, los efectos especiales fueron impresionantes",
    "No es mala, pero esperaba algo más emocionante",
    "Una obra maestra del cine español, increíble dirección",
    "El final fue confuso, no entendí nada",
    "Excelente cinematografía, pero el guion flojo",
    "Demasiado predecible, ya sabía lo que iba a pasar",
    "Me reí mucho, muy divertida y entretenida",
    "La música fue espectacular, pero los actores no convencieron",
    "Película mediocre, no aporta nada nuevo",
    "Excelente historia, emocionante hasta el final",
    "No me gustó la ambientación, parecía de bajo presupuesto",
    "Muy buena dirección y fotografía, la recomiendo",
    "El guion tenía agujeros, pero los efectos visuales impresionan",
    "Aburrida, diálogo poco natural",
    "Gran actuación de los protagonistas, realmente me impactó",
    "El ritmo es lento, pero la historia es interesante",
    "No la recomiendo, desperdicié mi tiempo",
    "Un clásico moderno, me encantó cada escena",
    "Las escenas de acción fueron espectaculares",
    "El humor es pobre y los personajes poco creíbles",
    "Me emocioné, la historia me llegó al corazón",
    "La trama es confusa y difícil de seguir"
]
```

```python
# Creación de CountVectorizer sin Procesamiento
vectorizer_count  =  CountVectorizer()
X_count  =  vectorizer_count.fit_transform(corpus_es)
```

```python
# Creación de TfidfVectorizer sin Procesamiento
vectorizer_tfidf =  TfidfVectorizer()
X_xfidf =  vectorizer_tfidf.fit_transform(corpus_es)
```

```python
# Impresión de resultados
print("---Ejercicio 1 ---")
print(f"Dimensiones de CountVectorizer: {X_count.shape}")
print(f"Dimensiones de TfidfVectorizer: {X_xfidf.shape}")
```

```python
# Impresión de los primeros 10 tokens con sus respectivos índices
tokens_count  =  list(vectorizer_count.vocabulary_.items())[:10]
print(f"\nPrimeros 10 tokens (Count): {tokens_count}")
```

```python
# Diferencia entre ambas tablas
```

```python
# Obtención dle nombre de las palabras
tokens  =  vectorizer_count.get_feature_names_out()

df_count  =  pd.DataFrame(X_count.toarray(), columns  =  tokens)
df_tfidf =  pd.DataFrame(X_xfidf.toarray(), columns  =  tokens)

# Impresión de primeras filas
print("--- CountVectorizer ---")
print(df_count.head())

print("\n--- TFIDF ---")
print(df_tfidf.head())
```

<p align  =  "Justify "

Como se puede observar CountVectorizer es una matriz que muestra el número de veces que una palabra aparece en el documento. En cambio, TF-IDF muestra el peso (importancia relativa) de cada palabra en el documento.

# Ejercicio 2

```python
nltk.download('stopwords')
stopwords_español  =  stopwords.words('spanish')

# CountVectorizer con Procesamiento de Datos
vectorizer_count  =  CountVectorizer(
    lowercase =  True,
    strip_accents =  'unicode',
    stop_words =  stopwords_español
)
X_count  =  vectorizer_count.fit_transform(corpus_es)
```

```python
# TF-DF Vectorizer con Procesamiento
vectorizer_tfidf = TfidfVectorizer(
    lowercase =  True,
    strip_accents = 'unicode',
    stop_words =  stopwords_español
)
Xtfidf =  vectorizer_tfidf.fit_transform(corpus_es)
```

```python
# Impresión de resultados
print("---Ejercicio 1 ---")
print(f"Dimensiones de CountVectorizer: {X_count.shape}")
print(f"Dimensiones de TfidfVectorizer: {Xtfidf.shape}")

# Impresión  de primeros 10 tokens
vocabulario  =  vectorizer_count.get_feature_names_out()
print(f"\nPrimeros 10 tokens: {vocabulario[:10]}")
```

```python

# Comparación de resultados
df_comparativo =  pd.DataFrame({
    'Token': vocabulario,
    'Count': X_count.toarray()[0],
    'TF-IDF': Xtfidf.toarray()[0]
})

print(df_comparativo[df_comparativo['Count'] > 0])
```

La principal diferencia es la reduccción del numero de dimensiones en el documento (124 vs 94). Adicionalmente, al eliminar stopwords se observa que se conservan tokens con mayor caracter semantica. Finalmente, se observa que la frencuencia realtiva de cada palabra se vuelve más precisa al eliminar stopwords.
