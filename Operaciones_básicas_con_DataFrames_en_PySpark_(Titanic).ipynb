```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, when, isnan

# Creación de sesión en Spark
spark = SparkSession.builder \
    .appName('Tarea Titanic: Operaciones Básicas') \
    .master('local[*]') \
    .getOrCreate()
```

Se configura la  sesión para poder interatuar con dataframes en Spark.

#1) Carga del Dataset

```python
df =  spark.read.csv('/content/drive/MyDrive/Ingeniería de Datos Avanzada/Titanic-Dataset.csv', header =  True, inferSchema  =  True)
```

#2) Exploración Inicial

```python
# Visualización de las primeras 5 filas
df.show(5)
```

```python
# Visualización del tipo de esquema el dataframe
df.printSchema()
```

<p align = "Justify"

Se observa que el esquema permite valores nulos para todas las columnas. De igual modo, se observa que variables categoricas como Embarked, Cabin, Sex están categorizadas como texto, lo cual requiere transformación para poder ser utilizadas en modelos de Machine Learning.

# 3) Selección  de columnas

```python
df_seleccion =  df.select("Name", "Age", "Sex")
df_seleccion.show(5)
```

Se realiza una selección de columnas para trabajar con un subconjunto de los datos.

#4) Filtrado de Datos

```python
df_filtered =  df.filter((col("Age") > 18) & (col("Sex") == "female"))
df_filtered.show(5)
```

Se realiza un filtrado para obtener a las mujeres mayores de 18 años.

#5) Valores Únicos

```python
print("Valores Únicos de la Columna Pclass ")
df.select("Pclass").distinct().show()
```

#6)Agrupación

```python
print("Pasajeros por Clase:")
df.groupBy("Pclass").count().show()
```

```python
print("Pasajeros por Sexo:")
df.groupby("Sex").count().show()
```

Se observa una mayor cantidad de pasajeros en la clase 3 con 491 registros contra 216 de la clase 1 y 184 de la clase 2. De igual modo, se observa una mayor cantidad de hombres en la muestra de datos.

# 7) Estadística Básica

```python
df.describe().show()
```

Se muestra  un analisis de estadistica descripva del conjunto de datos, por ejemplo, el 38.38% sobrevivió, la edad promedio es 29.69, ticket promedio de 260138.54, etc.

# 8 ) Valores Nulos

```python
df.select([
    count(when(isnan(c) | col(c).isNull(), c)).alias(c)
    if t in ("double", "float")
    else count(when(col(c).isNull(), c)).alias(c)
    for c, t in df.dtypes
]).show()
```

Se observa 177 registros sin edad y 687 sin registros en la columna Cabin, los cuales podrían ser remplazados por el valor medio o la moda.
