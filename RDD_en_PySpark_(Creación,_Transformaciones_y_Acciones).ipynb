```python
# Importación de PySpark
try:
    import pyspark
except ModuleNotFoundError:
    !pip install pyspark -q
    import pyspark

# Importacion de librerías básicas
import math
import numpy as np
from pyspark.sql import SparkSession
from pyspark import SparkContext

# Creación de SparkSession

sc = SparkContext.getOrCreate()
```

Se crea un objecto SparkContext para crear RDD y ejecutar operaciones distribuidas.

#1)

```python
np.random.seed(18)
lista_normal  =  np.random.normal(loc = 50, scale  =  10, size  =  1000 )
```

Se genera una lista de 1000 valores con media = 50, desviación estándar =  10, utilizando una semilla  =  18 para garantizar reproducibilidad.

#2)

Paraleliza la lista para crear un RDD llamado rdd0

```python

rdd0 = sc.parallelize(lista_normal)
```

Se paraleliza la lista de números aleatorios para realizar procesamiento distribuido.

# 3)

### A) Transformaciones

```python
rdd_cuadrado  =  rdd0.map(lambda x: x ** 2)
print("---Impresión de Primerios 5 Números al Cuadrado---\n")
print(rdd_cuadrado.take(5))
print()
print("---Impresión 4 Valores al Cuadrado Mayores a 50---\n")
rdd_filtrado = rdd_cuadrado.filter(lambda x:  x> 50)
print(rdd_filtrado.take(4))
```

Una vez creado el RDD se utiliza la función map() para elevar cada elemento al cuadrado. Con la función take() visualizamos los primeros 5 elementos y a través de la función filter() filtramos los elementos mayores a 50 y volvemos a visualizar los primeros 4 resultados usando la función take().

###B) Transformaciones en Pseudo-Conjuntos

```python
lista_normal_2 =  np.random.normal(loc= 50, scale  =  10, size  =  1000)
lista_normal_3 =  np.random.normal(loc  =  50, scale  =  10, size  = 1000)

rdd1 =  sc.parallelize(lista_normal_2)
rdd2 =  sc.parallelize(lista_normal_3)
```

De la misma manera se crean 2 nuevas listas de 1000 numeros aleatorios con media = 50, varianza =  10, y se crean los RDD utilizando la función parallelize() para operaciones distribuidas.

```python
# Transformaciones
rdd_union  =  rdd1.union(rdd2)
rdd_distinct =  rdd_union.distinct()
rdd_intersection =  rdd1.intersection(rdd2)
rdd_subtract =  rdd1.subtract(rdd2)
rdd_cartesian =  rdd1.distinct().cartesian(rdd2)
print("---Operaciones Registradas en el DAG de PySpark")
```

Posteriomente, se registran en el DAG de PySpark las siguientes tranformaciones:


*   Union la cual combina los elementos de los 2 RDDs
*   Distinct, la cual devuelve los elementos no repetidos  entre los 2 RDDs.
*   Intersection, la cual devuelve los elementos en común entre los 2 RDDs
*   Substract, la cual devuelve los elementos del primer RDD que no están contenidos en el segundo RDD.
*   Finalmente, cartesian devuelve todas las combinaciones posibles de pares entre los 2 RDDs (se utiliza previamente distinct para evitar combinacioens repetidas).







```python
# Ejecución de Accciones
print(f"Elementos Totales RDD Union: {rdd_union.count()}")
print(f"Muestra RDD Union: {rdd_union.take(5)}")
print()

print(f"Elementos Totales RDD Distinct: {rdd_distinct.count()}")
print(f"Muestra RDD Distinct: {rdd_distinct.takeOrdered(5)}")
print()

print(f"Elementos Totales RDD Intersection: {rdd_intersection.count()}")
print(f"Muestra RDD Intersection: {rdd_intersection.takeOrdered(5)}")
print()

print(f"Elementos Totales RDD Subtract: {rdd_subtract.count()}")
print(f"Muestra RDD Substract: {rdd_subtract.takeOrdered(5)}")
print()

print(f"Elementos Totales RDD Cartesian: {rdd_cartesian.count()}")
print(f"Muestra RDD Cartesian: {rdd_cartesian.takeOrdered(5)}")
```

Se ejecutan las acciones utilizando count() para visualizar el tamaño de cada RDD resultante y takeOrdered() para visualizar los primeros 5 elementos de cada transformación.

#4)

Utiliza rdd0 y calcula el promedio, valor máximo y mínio.
Calcula la varianza y desviación estándar
Utiliza la biblioteca math y calcula la raíz cuadrada. Imprime solo 5 resultados.

```python
rdd_promedio  =  rdd0.mean()
print(f"Promedio: {rdd_promedio}")

rdd_maximo =  rdd0.max()
print(f"Máximo: {rdd_maximo}")

rdd_minimo  =  rdd0.min()
print(f"Mínimo: {rdd_minimo}")

rdd_varianza =  rdd0.variance()
print(f"Varianza: {rdd_varianza}")

rdd_desviación =  rdd0.stdev()
print(f"Desviación estándar: {rdd_desviación}")

rdd_raiz =  rdd0.map(lambda x: math.sqrt(abs(x)))
print(f"Impresion de la raíz de los primeros 5 elementos: {rdd_raiz.take(5)}")
```

Finalmente se ejecutan las acciones de operaciones artimeticas usando la función map(). En este caso, las acciones son ejecutadas sin pasos intermedios para devolver el valor final. Por el contrario, al extraer la raiz cuadrada del RDD se ejecuta una transformación perezosa la cual es ejecutada al extraer los primeros 5 elementos utilizando la función take().
